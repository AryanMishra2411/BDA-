```python
from collections import defaultdict
from multiprocessing import Pool

# Sample input data
input_data = [
    "Deer Bear River",
    "Car Car River",
    "Deer Car Bear"
]

# Step 1: Splitting - Split each row into words
def split_words(row):
    return row.split()

# Step 2: Mapping - Map words to their counts
def map_words(words):
    word_counts = defaultdict(int)
    for word in words:
        word_counts[word] += 1
    return word_counts

# Step 4: Reducing - Reduce and aggregate word counts
def reduce_counts(word_counts_list):
    total_counts = defaultdict(int)
    for word_counts in word_counts_list:
        for word, count in word_counts.items():
            total_counts[word] += count
    return total_counts

# Main MapReduce process
def map_reduce(input_data):
    # Splitting
    words_list = Pool().map(split_words, input_data)

    # Mapping
    word_counts_list = Pool().map(map_words, words_list)

    # Shuffling (not explicitly needed in this example)

    # Reducing
    final_word_counts = reduce_counts(word_counts_list)

    return final_word_counts

# Execute the MapReduce process
if __name__ == '__main__':
    final_word_counts = map_reduce(input_data)

    # Final output
    for word, count in final_word_counts.items():
        print(f"{word}: {count}")
```
